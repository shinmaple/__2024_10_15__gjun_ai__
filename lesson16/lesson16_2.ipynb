```python
import pandas as pd
df1 = pd.read_csv('111各鄉鎮市區人口密度new.csv')
display(df1)
df1.style   #把型別變成styler<--視覺化的DataFrame
type(df1.style)
#Styler.format(formatter=None, subset=None, na_rep=None, precision=None, decimal='.', thousands=None, escape=None, hyperlinks=None)
df1.style.format(precision=2,thousands=',', decimal='.')

#Styler.format_index(formatter=None, axis=0, level=None, na_rep=None, precision=None, decimal='.', thousands=None, escape=None, hyperlinks=None)
def format1(value:int):
    return f"第{value}號"
df1.style\
.format(precision=2,thousands=',', decimal='.')\
.format_index(format1)
```

```python
df1 = pd.read_csv('111各鄉鎮市區人口密度new.csv')
display(df1.style)
df1.style\
.format(precision=2,thousands=',', decimal='.')\
.format_index(lambda value:f'第{value}號')          #lambda會自動return但只能一行,所以可以不用寫def那兩行了
```

```python
import numpy as np
weather_df = pd.DataFrame(np.random.rand(10,2)*5,index=pd.date_range(start="2021-01-01",periods=10),
                        columns=["Tokyo", "Taipei"])
weather_df
```

```python
weather_df.index    #dtype='datetime64[ns]'是timestamp
```

```python
   # %A代表完整星期、%a是簡化的星期
weather_df.style\
.set_caption('Weather conditions')\
.format_index(lambda v:v.strftime("%A"))\
.background_gradient(vmin=1,vmax=5,cmap='Purples')
```
