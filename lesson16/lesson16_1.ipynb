```python
import pandas as pd

df1 = pd.read_csv('111各鄉鎮市區人口密度.csv')
df1
```

```python
#使用info()檢查dataFrame
df1.info()  #查出來的Dtype=object代表是字串
```

```python
#實體方法rename()->改變欄位名稱
df2 = df1.rename(columns={
    'statistic_yyy':'統計年',
    'site_id':'區域別',
    'people_total':'年底人口數',
    'area':'土地面積',
    'population_density':'人口密度'
    })
df2
```

```python
#Drop()   DataFrame.drop(labels=None, *, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
df3 = df2.drop(index=0)
df3
```

```python
#dropna     DataFrame.dropna(*, axis=0, how=<no_default>, thresh=<no_default>, subset=None, inplace=False, ignore_index=False)
df4 = df3.dropna()
df4
```

```python
#DataFrame.replace(to_replace=None, value=<no_default>, *, inplace=False, limit=None, regex=False, method=<no_default>)
df4[-5:].values
df5 = df4.replace('… ',value=0)
df5.info()
```

```python
#DataFrame.astype(dtype, copy=None, errors='raise')
df6 = df5.astype({
    '年底人口數':'int',
    '人口密度':'int',
    '土地面積':'float'
})
df6.info()
```

```python
#不會使用for in迴圈,因為不會得到索引編號
for value in df5['年底人口數']:
    #print(type(value))
    print(value)
```

```python
#map
def to_int(value:str)->int:
    try:
        #print(type(value))
        #print(value)
        return int(value)   #轉成series
    except:
        return 0
df5['年底人口數'] = df5['年底人口數'].map(to_int)
```

```python
df5.info()
```

```python
df6 = df5.astype({'土地面積':'float','人口密度':'int'})
df6.info()
```

```python
df6
```

```python
#儲存成為csv檔
df6.to_csv('111各鄉鎮市區人口密度new.csv',index=False)
```

```python
#儲存成為ecxel檔
df6.to_excel('111各鄉鎮市區人口密度new.xlsx')
```

```python
#儲存成為json檔
df6.to_json('111各鄉鎮市區人口密度new.json',orient='records',force_ascii=False)
```
