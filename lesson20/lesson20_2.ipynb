## ticks,labels,legends
- xlim,xticks,xticklabels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)             #實體方法
ax.plot(np.random.randn(1000).cumsum()) #實體方法
#ax.set_xticks([0,250,500,750,1000])
# _ = ax.set_xticks([0,250,500,750,1000])     #_ =   <--是省略運算子
_ = ax.set_xticks([0,100,200,300,400,500,600,700,800,900,1000]) 
_ =ax.set_xticklabels(['zero','one','two','there','four','five','sex','seven','eight','nine','ten'])
ax.set_xlim(300,600)
ax.set_ylim(-20,50)
```
