```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#np.random.randn(10,4).cumsum()
#np.random.randn(10,4).cumsum(axis=0)
df = pd.DataFrame(np.random.randn(10,4).cumsum(axis=0),
             columns=['A','B','C','D'],
             index=np.arange(0,100,10)
             )

df.plot.line()
```

```python
fig, axes = plt.subplots(2,1)
series = pd.Series(np.random.rand(16),
          index=list('abcdefghijklmnop')          
          )
series.plot.bar(ax=axes[0],color='b',alpha = 0.7)
series.plot.barh(ax=axes[1],color='b',alpha = 0.7)
```

```python
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'],tips['size'])
#party_counts
party_counts1 = party_counts.loc[:,2:5]
party_pcts = party_counts1.div(party_counts.sum(axis=1),axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```
