```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
data = pd.read_csv('spx.csv',index_col=0,parse_dates=True)  #parse_dates,datetimeIndex
series = data['SPX']
fig = plt.figure(figsize=(10,5))
axes = fig.add_subplot(1,1,1)
crisis_data = [
    (datetime(2007,10,11),'Peak of bull market'),
    (datetime(2008,3,12),'Bear Stearns Fails'),
    (datetime(2008,9,15),'Lehman Bankruptcy')
]
for date, label in crisis_data:
    axes.annotate(label,
                  xy=(date,series[date]+75),
                  xytext=(date,series[date]+225),
                  arrowprops=dict(facecolor='red',headwidth=4,width=2,headlength=4),
                  horizontalalignment='left',
                  verticalalignment='top'
                  )
series.plot.line(ax=axes,style='k-')
axes.set_title('Important dates in the 2008-2009 financial crisis')
axes.get_xticks()

#axes.set_xlim('2007-01-01','2011-01-01')
axes.set_xlim(datetime(2007,1,1),'2011-01-01')
axes.set_ylim([600,2000])
```
