```python
onestore = np.random.randint(100,10000,size=(30,))
onestore
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,5))
axes1 = fig.add_subplot(2,1,1)
axes1.plot(range(1,31),onestore,'b-')
axes1.set_xticks(range(1,31))
axes2 = fig.add_subplot(2,1,2)
axes2.plot(range(1,31),onestore.cumsum(),'y-')
axes2.set_xticks(range(1,31))
plt.show()
```

```python
fivestore = np.random.randint(100,10000,size=(5,30))
fivestore_cumsum = fivestore.cumsum(axis=1)
fivestore_cumsum
```

```python
for one in fivestore_cumsum:
    print(one)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,5))
axes=fig.add_subplot()
axes.set_xticks(range(1,31))
axes.grid(axis='x')
axes.grid(axis='y')
for one in fivestore_cumsum:
    axes.plot(range(1,31),one)
plt.show()
```
